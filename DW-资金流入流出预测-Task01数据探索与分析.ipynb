### 赛题简介

蚂蚁金服拥有上亿会员并且业务场景中每天都涉及大量的资金流入和流出，面对如此庞大的用户群，资金管理压力会非常大。在既保证资金流动性风险最小，又满足日常业务运转的情况下，精准地预测资金的流入流出情况变得尤为重要。此届大赛以《资金流入流出预测》为题，期望参赛者能够通过对例如余额宝用户的申购赎回数据的把握，精准预测未来每日的资金流入流出情况。对货币基金而言，资金流入意味着申购行为，资金流出为赎回行为。

希望给出的预测方案在每一天的预测误差都较小。

### 学习内容

时序数据挖掘，通过挖掘2013-2014年度资金流入流出情况，预测未来每日资金流入流出。

在一般数据挖掘工作（挖掘自变量与因变量之间的相关关系）以外，时序数据挖掘有其自身特点：

1. 周期性。排除其他变量（利率变化等）因素后，数据波动可以解释为年、季度、月、星期等为单位周期性地变化的叠加。

2. 异常点。部分异常点可归结为节假日等特殊时间节点对周期性数据的影响。

### 赛题数据

竞赛提供了余额宝收益计算方式，以及余额宝用户的申购赎回数据(2013年7月-2014年8月;每天;28041位用户，2840421条记录)等信息。分为五张表：

1. 用户信息表:user_profile_table。分析不同用户的操作习惯。
2. 用户申购赎回数据表:user_balance_table。主数据表，已知的因变量部分。
3. 收益率表:mfd_day_share_interest。余额宝的浮动收益率表，影响用户的申购赎回意愿。
4. 上海银行间拆放利率表:Shibor。与表3类似。作为比对参考。需判断该利率与申购赎回意愿之间相关性。
5. 选手提交结果表(示例): tc_comp_predict_table。


### 数据探索与分析

#### 数据准备

比赛数据比较完整，格式统一，不需要进一步清洗。

赛题数据表中的时间戳信息以日为单位，字符串格式。需要格式转化，为数据集添加时间戳信息。

#### 时序分析

对每日的申购/赎回总量在不同时间尺度（年、月、周等）做时序图，分析数据周期性。14年4月之后数据较为平稳

#### 翌日（weekday）特征分析

- 每日的申购/赎回总量按weekday求和，观察星期X与申购赎回总额间关系。

- 分别做Spearman相关分析。

- 调用mvtest测试（待研究）翌日特征与标签的独立性。

#### 月份特征分析

- 整体：使用kdeplot(核密度估计图)按月对每日总量做估计
- 附近：对临近几个月做单独分析，比对各月kde分布异同

#### 日期特征分析

- 按月对每日申购/赎回均值画直方图，找周期规律。各月表现不同。13年9月由于周末分布及节假日情况无按星期周期性特征。
- 按每日画热力图，观察异常点（与附近点颜色差别较大），并反推确定原因。发现调休后的周末工作日以及利率大幅波动。

#### 对于节假日及周边日期的分析

- 观察节假日及其前后几日的申购赎回波动。

#### 对异常值的分析

- 对所有交易画箱形图，发现特大成交并对其分析
- 将交易量划分为大/小两组，分别统计规律

#### 分析用户交易纪录表中其他变量

对表2中所有变量做交叉商。分析两两相关性。

#### 银行及支付宝利率水平与交易关系

- 各种利率与申购赎回总量的时序变化图
- 各种利率变化与申赎之间相关性

#### 用户层面深入分析

- 分析14年4月之后大小额（以单次超过100w交易额为标准）用户的交易规律
- 对比分析高低频交易用户、新老用户的交易习惯
- 根据表1（用户信息），分析城市、性别、星座等对用户交易习惯的影响

#### 其他分析

针对余额宝的交易特性，对：

- 日利息与购买量；
- 支付宝利率与利息；
- 支付宝利率与【利息/购买量】；
- 不同申购/赎回方式（支付宝/银行账户）

进行时序图分析。

### 学习心得
对于时序预测问题，数据探索与分析工作，就是：
1. 对涉及到的所有参数进行不同时间尺度上的时序图（热力图）分析，试图找到某个时段，使得参数与交易量之间有较强的同步性；
2. 从各表出发，对可变参数（利率、利息等）做相关分析，对所有参数（性别、城市等）都可做分类标准研究交易量规律。
3. 使用热力图、箱型图等手段找到异常值，并通过额外的信息收集验证异常值产生原因，并在预测中予以考虑。
